```python
def find_urls_with_keyword(text, keyword=None):
    # 更新正则表达式以匹配 URL，直到遇到引号或空格
    url_pattern = r'https?://[^\s"]+'
    # 查找所有匹配的 URL
    urls = re.findall(url_pattern, text)

    # 过滤出包含关键词的 URL
    if keyword:
        filtered_urls = [url for url in urls if keyword in url]

    return filtered_urls
```

```python
def find_minimum_keys(dict):
    # 创建一个集合包含所有唯一的值
    all_values = set(val for values in dict.values() for val in values)
    selected_keys = []
    covered_values = set()

    while covered_values != all_values:
        # 选择一个包含最多未覆盖元素的键
        best_key = None
        best_key_values = set()
        for key, values in dict.items():
            uncovered_values = set(values) - covered_values
            if len(uncovered_values) > len(best_key_values):
                best_key = key
                best_key_values = uncovered_values

        # 更新已覆盖的值和选择的键
        covered_values.update(best_key_values)
        selected_keys.append(best_key)

    return selected_keys

# 测试算法
A = {'a': [1, 2, 3], 'b': [3, 4, 5], 'c': [1, 3, 4]}
print(find_minimum_keys(A))
```
